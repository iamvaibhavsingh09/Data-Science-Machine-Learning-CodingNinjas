### Square Patterns

Print the following pattern for the given N number of rows.
Pattern for N = 4

```python
n = int(input())
i = 1
while n>=i:
    j = 1
    while j<=n:
        print(i,end='')
        j = j+1
    print()
    i = i + 1
```

```python
n = int(input())
i = 1
while n>=i:
    j = 1
    while j<=n:
        print(j,end='')
        j = j+1
    print()
    i = i + 1
```

```python
n = int(input())
i = 1
while n>=i:
    j = 1
    while j<=n:
        print(n-j+1,end='')
        j = j+1
    print()
    i = i + 1
```
