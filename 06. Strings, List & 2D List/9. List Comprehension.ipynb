### List Comprehension

```python
li = [1,2,3,4]
li
```

```python
li_new =[]
for ele in li:
    li_new.append(ele**2)
    
print(li_new)
```

```python
li_n =[ele**2 for ele in li]
li_n
```

```python
li_e_sq =[ele**2 for ele in li if ele%2==0]
li_e_sq
```

```python
li = [1,2,3,4,5,6]
li_23 =[ele for ele in li if ele%2 ==0 if ele%3==0]
print(li_23)
```

```python
li_1 = [1,2,3,4,5]
li_2 = [2,4,6,7]

li_inter =[]
for ele in li_1:
    for ele_2 in li_2:
        if ele==ele_2:
            li_inter.append(ele)
print(li_inter)
```

```python
li_1 = [1,2,3,4,5,7]
li_2 = [2,4,6,7]
li_inter =[ele for ele in li_1 for ele_2 in li_2 if ele ==ele_2]
print(li_inter)
```

```python
li =[1,2,3,4,5]
li_e= [ele**2 if ele %2 ==0 else ele for ele in li]
li_e
```

```python
s = 'Parikh'
li = [ele for ele in s]
li
```
