```python
# Python assume we wnat to create  local variable.
a5 = 13
def f5():
    a5 = 12
    print(a5)
    print(id(a5))
print(a5)
f5()
print(a5)
print(id(a5))
```

```python
# Adding global 
a6 = 13
def f6():
    global a6
    a6 = 12
    print(a6)
    
print(a6)
f6()
print(a6)
```
