```python
n = int(input())
i = 1
while n>=i:
    j = 1
    while j<=n:
        charP =chr(ord('A') +j-1)
        print(charP,end='')
        j = j+1
    print()
    i = i + 1
```

```python
n = int(input())
i = 1
while n>=i:
    j = 1
    start_char = chr(ord('A')+i-1)
    while j<=n:
        charP =chr(ord(start_char) +j-1)
        print(charP,end='')
        j = j+1
    print()
    i = i + 1
```
